```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import numpy as np
import seaborn as sb
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
!wget https://raw.githubusercontent.com/deaashilla/midml/main/mydataset.csv
```

Load the profile of the dataset

```python
mydataset = pd.read_csv("mydataset.csv")
print(mydataset.dtypes)
print(mydataset.isna().values.any())
```

making new data frame that i have selected to create the linear regression 

```python
linmydataset = mydataset[["number_of_reviews", "longitude", "price", "reviews_per_month", "minimum_nights"]]
print(linmydataset.head())
```

training the data using linear regression 

```python
train, test = train_test_split(linmydataset, test_size=0.2)
regression = linear_model.LinearRegression()
regression.fit(train[["number_of_reviews"]], train[["minimum_nights"]])
print('coefficients: ', regression.coef_)
print('intercept: ',regression.intercept_)
```

then we evaluate the data by predicting the linear regression

```python
prediction = regression.predict(test[["number_of_reviews"]])
for i in range(len(test)):
  print(test[["number_of_reviews"]].values[i], prediction[i])

print("mean absolute error : ", mean_absolute_error(test[["minimum_nights"]], prediction))
print("mean squared error : ", mean_squared_error(test[["minimum_nights"]], prediction))
print("R2 : ", r2_score(test[["minimum_nights"]], prediction))
```
